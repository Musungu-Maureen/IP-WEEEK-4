```python
# Importing the relevant libraries to our environment
import pandas as pd
import numpy as np
```

```python
# Loading our datasets
Autolib = pd.read_csv('Autolib_dataset.csv')
```

```python
# Displaying our dataframe
Autolib.head()
```

```python
# Validity: Checking for any irrelevant data
# The geological coordinates of the station will not help us in any way in our study
#We therefore drop the 'Geo Point' attribute.
Autolib1 = Autolib.drop('Geo point', axis=1)
Autolib1.head()
```

```python
#Deleting the attribute 'Displayed Comment'
Autolib2 = Autolib1.drop('Displayed comment', axis=1)
Autolib2.head()
```

```python
#Checking for missing values 
Autolib2.isnull().sum()
#Explanation:There are 4953 missing values in the column 'Scheduled at'. 
#We will therefore drop this column since a large percentage of it is already null
Autolib3 = Autolib2.drop('Scheduled at', axis=1)
Autolib3.head()
```

```python
#Checking for duplicates in the data
Autolib3.duplicated().any()
#There are no duplicates in our data so we are good to go!
Autolib3.info()
```

```python
# Exporting our final dataset to csv so we can use it for analysis in SQL
Autolib3.to_csv("Final_Autolib.csv")

```
